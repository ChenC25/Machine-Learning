```python
cars = cars.drop(['VRVIN'], axis=1)
cars = cars.drop(['Vehicle_MMR'], axis=1)
cars.head()
```

```python
# split into X and y
X = cars.loc[:, ['VRMILEAGE', 'Vehicle_year',
       'Vehicle_cylinders', 'Vehicle_doors', 'Vehicle_engine', 'Vehicle_condition_overall',
       'Vehicle_condition_grade_1', 'Vehicle_condition_grade_2', 'VRSELLTYPE', 'Location_LocationName', 'Vehicle_make',
       'Vehicle_model', 'Vehicle_botcolor', 'Vehicle_intcolor', 'Vehicle_inttype', 'Vehicle_topcnstr',
       'Vehicle_trantype', 'Vehicle_fuel', 'Vehicle_btext', 'Vehicle_bshort', 'Vehicle_drive',
       'Vehicle_Mileage_Tier', 'Vehicle_Pricing_Tier', 'Vehicle_condition_crstatus', 'Vehicle_condition_drivable','Misc_SalesChannel']]

y = cars['VRSALEAMT']
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import category_encoders as ce
```

```python
# Label Encoding
le = LabelEncoder()
label_mappings = {}

for col in ['VRSELLTYPE', 'Vehicle_trantype', 'Vehicle_fuel', 'Vehicle_drive', 'Misc_SalesChannel','Vehicle_condition_drivable',
            'Vehicle_condition_crstatus', 'Vehicle_inttype', 'Vehicle_topcnstr']:
    cars_categorical[col] = le.fit_transform(cars_categorical[col])
    label_mappings[col] = dict(zip(le.classes_, le.transform(le.classes_)))

# Frequency Encoding
frequency_mappings = {}

for col in ['Vehicle_Mileage_Tier', 'Vehicle_Pricing_Tier','Vehicle_make', 'Vehicle_model', 
            'Vehicle_botcolor', 'Vehicle_intcolor', 'Location_LocationName', 'Vehicle_btext', 'Vehicle_bshort']:
    freq_map = cars_categorical[col].value_counts().to_dict()
    cars_categorical[col] = cars_categorical[col].map(freq_map)
    frequency_mappings[col] = freq_map

# Handling Missing Values
cars_categorical['Vehicle_condition_crstatus'].fillna('Unknown', inplace=True)
cars_categorical['Vehicle_condition_drivable'].fillna('Unknown', inplace=True)

# # Target Encoding
# target_enc = ce.TargetEncoder()
# for col in ['Vehicle_make', 'Vehicle_model', 'Vehicle_botcolor', 'Vehicle_intcolor', 'Location_LocationName', 'Vehicle_btext', 'Vehicle_bshort']:
#     cars_categorical[col] = target_enc.fit_transform(cars_categorical[col], cars_categorical.index) 
 
cars_categorical.head()
```

```python
# drop categorical variables 
X = X.drop(list(cars_categorical.columns), axis=1)
```

```python
# concat dummy variables with X
X = pd.concat([X, cars_dummies], axis=1)
```

```python
# scaling the features
from sklearn.preprocessing import scale

# storing column names in cols, since column names are (annoyingly) lost after 
# scaling (the df is converted to a numpy array)
cols = X.columns
X = pd.DataFrame(scale(X))
X.columns = cols
X.columns
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import train_test_split
```

```python
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    train_size=0.95,
                                                    test_size = 0.05, random_state=42)
```

```python
# Instantiate Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf.fit(X_train, y_train)

# Print feature importances
print(rf.feature_importances_)
```

```python
# Predict
y_pred = rf.predict(X_test)

# Calculate r2 score
print(r2_score(y_true=y_test, y_pred=y_pred)) 
```
